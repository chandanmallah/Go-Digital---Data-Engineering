                                    #Task 6
                                    
  >   Business wants to check the Claim Amount for deviation for each transaction, they would like you to calculate as follow 
   
  >  CLAIM_DEVIATION = AVG_CLAIM_AMOUNT_FOR_LAST_30DAYS (same insurance type) / CURRENT_CLAIM_AMOUNT 
  
  >  If the value < 0.5 THEN CLAIM_DEVIATION = 1 ELSE 0
  
  >  If there is less than 30 days of transaction history THEN -1 Note: LAST_30DAYS does not include current day

```python
merged.TXN_DATE_TIME.max()
```

```python
# Converting TXN_DATE_TIME feature max value into a datetime form
# Assuming current day to be 2021-06-30
todays_date = pd.to_datetime('2021-06-30')
todays_date
```

```python
merged['TXN_DATE_TIME'] = pd.to_datetime(merged['TXN_DATE_TIME'])
```

```python
# df with last 30 days data
last_30 = merged.loc[(merged['TXN_DATE_TIME'] > '2021-05-30') & (merged['TXN_DATE_TIME'] < '2021-06-30')] 
```

```python
last_30
```

```python
# get average claim amount from last 30 days by grouping with INSURANCE_TYPE and converting it into a dictionary
avg_claim = last_30.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()
avg_claim
```

```python
# creating a function for calculating claim_deviation
def Claim_deviation(row):
    if (todays_date - row['TXN_DATE_TIME']) < pd.Timedelta(days=30):
        return -1
    claim_deviation = avg_claim[row['INSURANCE_TYPE']] / row['CLAIM_AMOUNT']
    if claim_deviation < 0.5:
        return 1
    else:
        return 0
```

```python
# Applying the function to the dataframe
merged['CLAIM_DEVIATION'] = merged.apply(Claim_deviation,axis=1)
```

```python
# getting the values for feature CLAIM_DEVIATION
merged['CLAIM_DEVIATION'].value_counts()
```

Answer : Values which are less than 0.5 there are 9216 claims whereas 784 claims with transaction history of less than 30 days.

> Business wants to check Claim Velocity which is defined as follow

> CLAIM_VELOCITY = NO_OF_CLAIMS_IN_LAST30DAYS (for the current insurance type)
/ NO_OF_CLAIMS_IN_LAST3DAYS (for the current insurance type)

> Note: LAST30DAYS & LAST3DAYS does not include current

```python
# Getting the current date value
todays_date
```

```python
# get number of values in feature 'CLAIM_STATUS'

merged['CLAIM_STATUS'].value_counts()
```

```python
# Get the values for last 3 days and converting it into timedelta, It is used represent the duration, the difference between two dates or times.
last_3_days = merged.loc[(merged['TXN_DATE_TIME']>todays_date - pd.Timedelta(days=4)) & (merged['TXN_DATE_TIME'] < todays_date)]
last_3_days
```

```python
# Calculating total approved claims insurance type wise in last 30 days
claims_in_last_30_days = pd.DataFrame.from_dict(last_30[last_30['CLAIM_STATUS']=='A'].groupby('INSURANCE_TYPE').size().to_dict(),orient='index',columns=['NO_OF_CLAIMS_IN_LAST30DAYS'])
claims_in_last_30_days
```

```python
# Calculating total approved claims insurance type wise in last 3 days
claims_in_last_3_days = pd.DataFrame.from_dict(last_3_days[last_3_days['CLAIM_STATUS']=='A'].groupby('INSURANCE_TYPE').size().to_dict(),orient='index',columns=['NO_OF_CLAIMS_IN_LAST3DAYS'])
claims_in_last_3_days
```

```python
# Fetch the value of approved claims by joining claims in last 30 and 3 days
approved_claims = claims_in_last_30_days.join(claims_in_last_3_days)
approved_claims
```

```python
# Create new feature CLAIM_VELOCITY by calculating difference between no. of claims in last 30 days and 3 days
approved_claims['CLAIM_VELOCITY'] = approved_claims['NO_OF_CLAIMS_IN_LAST30DAYS']/approved_claims['NO_OF_CLAIMS_IN_LAST3DAYS']
approved_claims['CLAIM_VELOCITY']
```

                                #Task 12
> Business wants to find all Suspicious Employees (Agents).

> IF TOTAL CLAIM AMOUNT which meet below criteria is >= 15000 THEN AGENT is classified
as Suspicious ELSE Not

> CLAIM_STATUS = Approved AND CUSTOMER_RISK_SEGMENTATION = High

> AND INCIDENT_SEVERITY = “Major Loss”

> If Suspicious, then 1 ELSE 0. Find mean of this column.

```python
# Creating criteria in which various conditions are applied for features
criteria = (merged['CLAIM_STATUS']=='A')&(merged['RISK_SEGMENTATION']=='H')&(merged['INCIDENT_SEVERITY']=='Major Loss')
```

```python
# transactions that fullfil criteria
criteria_fulfilled = merged[criteria]
criteria_fulfilled
```

```python
# Joining condition fulfilled with AGENT_ID feature
grouped_suspicion_data = criteria_fulfilled.groupby('AGENT_ID').sum()
```

```python
# storing agents_id that are suspicious(met all the criteria)
suspicious_agents = list(grouped_suspicion_data[grouped_suspicion_data['CLAIM_AMOUNT']>=15000].index)
```

```python
employee_data
```

```python
# Make values of SUSPICIOUS feature as 0
employee_data['SUSPICIOUS'] = 0
```

```python
# Fetch suspicious employee based on their AGENT_ID 
suspicious_employee_slice = employee_data['AGENT_ID'].isin(suspicious_agents)
suspicious_employee_slice
```

```python
# Changing suspicious feature value to 1 of suspicious employees
employee_data.loc[suspicious_employee_slice,'SUSPICIOUS']=1
employee_data['SUSPICIOUS']
```

```python
# Fetch the mean value for the feature 'SUSPICIOUS'
employee_data['SUSPICIOUS'].mean()
```
